```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

G = control.tf([5],[3,0, 4])
print(G)

y,t = control.step(G)
plt.plot(t,y)
plt.xlabel('Time')
plt.title('Step Response')
plt.grid()

#Transfer function adalah mencari hubungan antara input dan output
# STEP biar tahu respons
```

```python
Kp= 0.01
G = control.tf([Kp*5],[3,0, 4])

#Ki= 5
#Kd= 5
sys = control.feedback( G)
print(G)


y,t = control.step(sys)
plt.plot(t,y)
plt.xlabel('Time')
plt.title('Step Response')
plt.ylim([-2,5])
plt.grid()
```
